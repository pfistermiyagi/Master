```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
mass = pd.read_csv(r'C:\Users\AP\Downloads\Mass Shootings Dataset Version 5.csv')
```

```python
mass.columns
```

```python
# Drop NaN values in dataframe

mass.dropna(inplace=True)
```

```python
sns.violinplot(y='Avg Age', data=mass)
plt.title('Avg Age of Shooter')
```

```python
sns.boxplot(y='Avg Age', data=mass)
plt.title('Avg Age of Shooter')

print("Mean age: ", mass['Avg Age'].mean())
print("Median age: ", mass['Avg Age'].median())
```

```python
sns.kdeplot(mass['Avg Age'])
plt.title('Avg Age of Shooter')
plt.xlabel('Avg Age')
```

```python
sns.distplot(mass['Avg Age'], bins=10)
plt.title('Avg Age of Shooter')
```

```python
mass['Year'] = [np.int(i.split('/')[2]) for i in mass['Date']]
```

```python
mass['Incidents'] = mass['Year'].value_counts().sort_index(ascending=True)
```

```python
sns.relplot(x='Injured', y='Fatalities', data=mass)
plt.title('Fatalities vs Injured')
```
