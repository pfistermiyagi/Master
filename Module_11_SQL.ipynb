WITH
    rainy 
AS (
    SELECT 
   	    date(date) rain_date
    FROM
        weather
    WHERE 
        events = 'Rain'
    GROUP BY 1
) 

SELECT
	t.trip_id,
	t.duration,
	date(t.start_date)
FROM 
    trips t
JOIN 
    rainy r
ON 
    date(t.start_date) = r.rain_date
ORDER BY duration DESC
LIMIT 3

    rainy
AS (
    SELECT 
		date(date) weather_date
    FROM 
        weather
    WHERE 
        events = 'Rain'
    GROUP BY 1

WITH
    rain_trips
AS (
    SELECT
		trip_id,
		duration,
		date(trips.start_date) rain_trips_date
    FROM 
        trips
    JOIN 
        rainy
    ON
        rainy.weather_date = date(trips.start_date)
    ORDER BY duration DESC
)

SELECT 
	rain_trips_date,
	max(duration) max_duration
FROM 
    rain_trips
GROUP BY 1
ORDER BY max_duration DESC
