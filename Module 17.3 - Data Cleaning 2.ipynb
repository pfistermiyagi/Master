The columns Primary, State, and Year have zero null values so nothing needs to be done with them. The last four columns comprising average reading and math scores have around 65% of their values missing so we'll drop those columns. The rest of the columns are float numbers and have a small percentage of null values, so we can fill these in with a mean number from the rest of the column.

```python
df.columns
```

```python
df_1 = df.copy()

fill_list = ['ENROLL', 'TOTAL_REVENUE',
       'FEDERAL_REVENUE', 'STATE_REVENUE', 'LOCAL_REVENUE',
       'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE',
       'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE',
       'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G', 'GRADES_KG_G',
       'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G',
       'GRADES_9_12_G', 'GRADES_ALL_G']

for col in fill_list:
    df_1.loc[:, col].fillna(df_1.loc[:, col].mean(), inplace=True)
```

```python
df_1.isnull().sum()/df_1.isnull().count()
```

------------------------------------------------------

3. Now, take into account the time factor. Replicate your second answer but this time fill in the missing values by using a statistic that is calculated within the year of the observation. For example, if you want to fill a missing value for a variable with the mean of that variable, calculate the mean by using only the observations for that specific year.

```python
df_2 = df.copy()

years = df_2["YEAR"].unique()

for col in fill_list:
    for year in years:
        df_2.loc[df_2["YEAR"] == year, col].fillna(df_2[df_2["YEAR"] == year][col].mean(), inplace=True)
```

-----------------------------------------

4. This time, fill in the missing values using interpolation (extrapolation).

```python
df_3 = df.copy()

for col in fill_list:
    df_3.loc[:, col].interpolate(inplace=True)
```

--------------------------------------------------

5. Compare your results for the 2nd, 3rd, and 4th questions. Do you find any meaningful differences?

```python
for col in fill_list:
    print("Statistics: {}".format(col))
    print(pd.concat([df[col], df_1[col],
                     df_2[col], df_3[col]], axis=1).describe())
```

```python
fig = plt.figure(figsize=(20,100))

for i, col in enumerate(fill_list):
    df_total = pd.concat([df[col], df_1[col],
                     df_2[col], df_3[col]], axis=1)
    
    plt.subplot(len(fill_list), 4, i+1)
    plt.bar(["Original", "Mean", "Year", "Interpolation"], df_total.median())
    plt.title("Mod of {}".format(col))
    
plt.tight_layout()
plt.show()
```

It doesn't seem like there is a major difference between the different ways of filling the cells, but there is a pattern emerging depending on which solution one uses. Using the mean method produces higher results than the original. Using the year method is equal to the original. And using interpolation produces less results than original.
